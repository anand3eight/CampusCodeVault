* Data Visualization

```python
#Import matplotlib
import matplotlib.pyplot as plt
```

```python
#Plot two attributes : Provide title, label, legend
students.plot(x="id", y="mark")
plt.title("Mark Graph")
plt.xlabel("ID")
plt.ylabel("Marks")
plt.legend()
```

```python
#Add a grid to the plot
students.plot(x="id", y="mark")
plt.title("Mark Graph")
plt.xlabel("ID")
plt.ylabel("Marks")
plt.legend()
plt.grid()
```

```python
#Create Subplots
names1 = students.head()['name']
marks1 = students.head()['mark']
names2 = students.tail()['name']
marks2 = students.tail()['mark']
fig, (ax1, ax2) = plt.subplots(2)
ax1.bar(names1, marks1)
ax1.set_title("First 5 Students")
plt.subplots_adjust(hspace=0.5)
ax2.plot(names2, marks2)
ax2.set_title("Last 5 Students")
```

```python
#Create a pie chart
plt.pie(x=students.head()["mark"], labels=students.head()["name"])
plt.title("Marks of First 5 Students")
plt.show()
```

```python
#Draw Horizontal and Vertical Bar Graphs for an attribute
def bar_graph(x, y, X_LABEL = 'X', Y_LABEL = 'Y', TITLE = 'Bar Graph', flag = True) :
    fig, ax = plt.subplots()
    graph = ax.bar(x, y) if flag else ax.barh(x, y)
    ax.set_xlabel(X_LABEL)
    ax.set_ylabel(Y_LABEL)
    ax.set_title(TITLE)
```

```python
bar_graph(students.head()['name'], students.head()['mark'], 'Name', 'Mark', 'Marks Graph')
bar_graph(students.head()['name'], students.head()['mark'], 'Mark', 'Name', 'Marks Graph', False)
```

```python
#Plot Different Markers with Different Colors
markers = ['o', '*', '.', 'h', 'x', 'D', '+', 's']
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k', 'w']
for i in range(8) :
    plt.plot(students.iloc[i]["id"], students.iloc[i]["mark"], markers[i]+':'+colors[i])
```
