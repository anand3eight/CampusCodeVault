**15. Write a Pandas program to import excel data (employee.xlsx ) into a Pandas dataframe
and find a list of employees where hire_date&gt; 01-01-07.**

```python
#!pip install python-docx
```

```python
import docx
data = docx.Document('employee.docx')
```

```python
for i in data.paragraphs :
  print(i.text)
```

```python
finalData = []
ls = list()
for i in data.paragraphs :
  if i.text.startswith('Sheet') :
    finalData.append(ls)
    ls = list()
  else :
    cur = i.text.split('\t')
    if(len(cur) > 1) :
      ls.append(cur)
finalData.append(ls)
```

```python
finalData
```

```python
columns = finalData[1][0]
```

```python
columns
```

```python
finalData.pop(0)
finalData[0].pop(0)
```

```python
sheet1 = pd.DataFrame(finalData[0], columns=columns)
sheet1
```

```python
sheet2 = pd.DataFrame(finalData[1], columns=columns)
sheet2
```

```python
sheet3 = pd.DataFrame(finalData[2], columns=columns)
sheet3
```

```python
#!pip install xlsxwriter
```

```python
def dfs_tabs(df_list, sheet_list, file_name):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')
    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)
    writer.save()

# list of dataframes and sheet names
dfs = [sheet1, sheet2, sheet3]
sheets = ['1','2','3']

# run function
dfs_tabs(dfs, sheets, 'employee.xlsx')
```

**25. Write a Pandas program to import three datasheets from a given excel data
